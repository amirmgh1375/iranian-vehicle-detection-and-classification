```python
from keras.models import load_model
model = load_model('CarModel_final_7.h5')
```

```python
from keras.preprocessing import image
from matplotlib.pyplot import imshow
import numpy as np
%matplotlib inline
```

```python
def predict_classes(predict):
    index = np.where(predict[0] == max(predict[0]))[0][0]
    name_dic = {0:"206", 1:"206 SD", 2:"207", 3:"405", 4:"arisan", 5:"dena", 6:"mazda-2000", 7:"neysan", 8:"pars", 9:"peykan", 10:"pride", 11:"rana", 12:"rio", 13:"samand", 14:"sayna", 15:"soren", 16:"suzuki", 17:"tiba", 18:"tondar90"}
    return name_dic[index]
```

```python
# img_path = 'E:/dataset/catDog/catVsdog/test/cats/cat.1500.jpg'
img_path = './img/car28.jpg'

img = image.load_img(img_path, target_size=(300, 300))
imshow(img)
x = image.img_to_array(img)
x /= 255.
x = np.expand_dims(x, axis=0)
predict = model.predict(x)
# print(predict)
```

```python
predict_classes(predict)
```
